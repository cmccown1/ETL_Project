```python
filename = os.path.join('Data', 'Consumer_Complaints.csv')

complain_df = pd.read_csv(filename, encoding='ISO-8859-1')

complain_df.head()

```

```python
# Get the data for 2018
complain_df['Date received'] = pd.to_datetime(complain_df['Date received'], format='%m/%d/%Y')

complain_df['Year'] = complain_df['Date received'].apply(lambda x: x.year)

df_2018 = complain_df.loc[complain_df["Year"] == 2018]

df_2018.head()
```

```python
# Get the complain count by state
count_df = df_2018["State"].value_counts()
count_df
```

```python
count_df.to_sql('complaints',con=engine, if_exists='replace')
```
